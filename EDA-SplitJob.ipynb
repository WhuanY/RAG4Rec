```python
import re
def split_sent(text):
    text = re.split('(?:[0-9][.;。：．•）\)])', text)  # 按照数字分割包括  1.  1;  1。  1：  1) 等
    ans = []
    for t in text:
        for tt in re.split('(?:[\ ][0-9][、，])', t):  #
            for ttt in re.split('(?:^1[、，])', tt):   # 1、
                for tttt in re.split('(?:\([0-9]\))', ttt):   # (1)
                    ans += re.split('(?:[。；…●])', tttt)
    
    return [_.strip()[:20] for _ in ans if len(_.strip()) > 0] # 限制一个工作描述最多20个item

def clean_text(text):
    illegal_set = ',.;?!~[]\'"@#$%^&*()-_=+{}\\`～·！¥（）—「」【】|、“”《<》>？，。…：'   # 定义非法字符

    for c in illegal_set:
        text = text.replace(c, ' ')     # 非法字符 替换为 空格
    for pattern in ['岗位职责', '职位描述', '工作内容', '岗位描述', '岗位说明', '工作职责', '你的职责']:
        text = text.replace(pattern, '')   # 内容头部替换为空格
    text = ' '.join([_ for _ in text.split(' ') if len(_) > 0])
    return text    # 空格间隔

length = []
for sent in idoc_df["job_doc:token_seq"]:
    sent = clean_text(sent)
    sent = split_sent(sent)
    length.append(len(sent))
    if len(sent) > 20:
        print(len(sent), "    ", sent)

from collections import Counter
count = Counter(length)
sorted_count = sorted(count.items())

# 输出结果
for num, freq in sorted_count:
    print(num, freq)
```
